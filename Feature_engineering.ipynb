**1.Feature selection**

Some of the features cannot reasonably be expected to have predictive potential, then we can drop them

```python
#create a new dataframe named chrun_df that excludes RowNumber,CustomerId,Surname,Gender columns
churn_df=original_data.drop(['RowNumber','CustomerId','Surname','Gender'],axis=1)
churn_df.head(10)
```

**2.Feature extraction**

```python
# Let  us  assume that individuals   who have been customers for a greater proportion of their lives might be less likely to churn.
#Let us create  Loyalty feature to  indicate the percentage of each customer's life that they were customers

churn_df['Loyalty']=churn_df['Tenure']/churn_df['Age']
churn_df.head(5)
```

**3.Feature transformation**

```python
# This step transforms our features to get them ready for modeling. This task is done based on model employed. Let us assume we want to construct a classification model
# generally classsification models need categorical variables to be encoded, thus we need to encode our categorical variable viz 'Geography'
churn_df['Geography'].unique()
```

```python
# Dummy encode categorical variables
churn_df = pd.get_dummies(churn_df,drop_first=True)
churn_df.head(5)
```
